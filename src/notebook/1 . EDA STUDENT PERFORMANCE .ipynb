#### The Percentiles (25%, 50%, 75%)

Think of these as "rank markers." If you lined up all 1,000 students from lowest score to highest score:

- 25% (First Quartile): This is the "cutoff" for the bottom 25% of students. In your math scores, 25% of students scored 57.0 or lower.

- 50% (Median): This is the exact middle. 50% of students scored below 66.0, and 50% scored above it. 
Note: Notice your mean (66.08) and 50% (66.0) are almost identical. This tells an engineer that your data is very symmetrical (not skewed).

- 75% (Third Quartile): This is the cutoff for the top 25% of students. 75% of students scored 77.0 or lower.

#### Insight
- From above description of numerical data, all means are very close to each other - between 66 and 68.05;
- All standard deviations are also close - between 14.6 and 15.19;
- While there is a minimum score 0 for math, for writing minimum is high = 10 and for reading it is much higher = 17

#### Can I conclude that since standard deviation of writing score is greater than maths, the teacher for writing may be bad and not all students are following them? 

1. The Spread vs. The Quality: 
A higher std (15.19 for writing vs. 15.16 for math) technically means the writing scores are more diverse.
    Your Logic: If the std is high, it could mean the teacher is only teaching the "top" students, while the rest are left behind, creating a wide gap.
    The Counter-Argument: Writing is a highly subjective skill. Some students might naturally have a "flair" for it, while others struggle with grammar or creativity, regardless of the teacher. Math is often more structured, which can lead to more consistent (though not necessarily better) scores.

2. Check the Mean (Average): 
To judge the teacher, look at the Mean alongside the std:
    If the Writing Mean is much lower than the Math Mean, and the std is higher, your "bad teacher" theory gains a lot of weight.
    In your data, the Writing Mean is 68.05 and the Math Mean is 66.08.
    The Reality: Even though the writing scores are more spread out, the students are actually performing better on average in writing than in math. This might actually suggest the writing teacher is doing a decent job!

3. Detecting the "Following" (The Skewness): 
Look at the 50% (Median) vs the Mean:
    If a large group of students "stopped following" the teacher, you would see a left-skewed distribution (a long tail of very low scores).
    In your data, the Writing Mean (68.05) and Median (69.00) are very close. This suggests that the "middle" of the class is still very much in sync with the lessons.
